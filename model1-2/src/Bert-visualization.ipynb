```python
import os 
import json
import torch
import pickle
import pandas as pd 
import numpy as np 
from attrdict import AttrDict
from transformers import AutoTokenizer, AutoConfig, AutoModel, utils
from bertviz import model_view
from transformers import BertForSequenceClassification
from transformers import BertConfig, BertTokenizer, BertModel
```

```python
data_path = "D:/AuD/model1-2/data"
base_model_path = "D:/AuD/base-model"
save_model_path = "D:/AuD/model1-2/model/"
config_path = "D:/AuD/model1-2/config"
log_path = "D:/AuD/model1-2/log"
config_file = "bert-base.json"
```

```python
tokenizer = BertTokenizer.from_pretrained(os.path.join(base_model_path, 'bert-base'), model_max_length=32)
config = BertConfig.from_pretrained(os.path.join(base_model_path, 'bert-base', 'bert_config.json'), num_labels=10, output_hidden_states=True, output_attentions=True)
model = BertForSequenceClassification.from_pretrained(os.path.join(base_model_path, 'bert-base'), config=config)
```

```python
with open(os.path.join(config_path, 'training_config.json')) as f:
    training_config = AttrDict(json.load(f))

training_config.pad = 'max_length'
training_config.device = torch.device("cuda") if torch.cuda.is_available() else "cpu"
model.to(training_config.device)
```

```python
config.max_position_embeddings = 128
```

```python
model_name = os.path.join(save_model_path, 'label_10', 'bert-base', 'bert_dsm_1.pt')
```

```python
model.load_state_dict(torch.load(model_name, map_location=torch.device('cpu')))
model.to(training_config.device)
```

```python
input_text = "I am so sad these days [SEP] and I lost 15 lbs"  
# input_text = "I went to school in the morning"
inputs = tokenizer.encode(input_text, return_tensors='pt').to(training_config.device)
```

```python
len(inputs[0])
```

```python
outputs = model(inputs)  # Run model
attention = outputs[-1]  # Retrieve attention from model outputs
tokens = tokenizer.convert_ids_to_tokens(inputs[0])  # Convert input ids to token strings
model_view(attention, tokens)  # Display model view
```

```python
from bertviz import head_view

head_view(attention, tokens)
```
